Question 1 : Explain what the forEach method does in JavaScript arrays and write code
that uses forEach to display each student's name from an array of student objects to the
console.
// Array of student objects
const students = [
{ name: "Rahul", age: 20 },
{ name: "Priya", age: 19 },
{ name: "Amit", age: 21 },
{ name: "Sneha", age: 20 }
];


The forEach() method is used to iterate over each element in an array. It executes a provided function once for each array element. It does not return a new array.

```python
const students = [
  { name: "Rahul", age: 20 },
  { name: "Priya", age: 19 },
  { name: "Amit", age: 21 },
  { name: "Sneha", age: 20 }
];
students.forEach(student => {
  console.log(student.name);
});
```

Question 2 : You have an array of product prices in rupees. Write code using the map
method to create a new array that contains the same prices but with 18% GST added to
each price.
const originalPrices = [100, 250, 500, 750, 1000];


```python
const originalPrices = [100, 250, 500, 750, 1000];
const pricesWithGST = originalPrices.map(price => {
  return price + (price * 0.18);
});
console.log(pricesWithGST);
```

Question 5 : Explain how the sort method works in JavaScript and write code to sort an
array of employee objects by their salary in descending order (highest salary first).
// Array of employee objects
const employees = [
 { name: "Rajesh", salary: 45000 },
 { name: "Priya", salary: 55000 },
 { name: "Amit", salary: 40000 },
 { name: "Kavya", salary: 60000 }
];

The sort() method sorts array elements.
By default, it converts elements to strings and sorts alphabetically.
To sort numbers or objects properly, we provide a comparison function.

For descending order:

If b.salary - a.salary is positive â†’ swap order.

Highest salary comes first.

```python
const employees = [
  { name: "Rajesh", salary: 45000 },
  { name: "Priya", salary: 55000 },
  { name: "Amit", salary: 40000 },
  { name: "Kavya", salary: 60000 }
];
employees.sort((a, b) => b.salary - a.salary);
console.log(employees);
```

Question 6 : Write code using the reduce method to calculate the total value of all items
in a shopping cart. Each item has a price and quantity. Also find the most expensive
single item in the cart.
// Shopping cart with items
const cart = [
{ name: "Laptop", price: 50000, quantity: 1 },
{ name: "Mouse", price: 500, quantity: 2 },
{ name: "Keyboard", price: 1500, quantity: 1 },
{ name: "Monitor", price: 15000, quantity: 1 }
];

```python
const cart = [
  { name: "Laptop", price: 50000, quantity: 1 },
  { name: "Mouse", price: 500, quantity: 2 },
  { name: "Keyboard", price: 1500, quantity: 1 },
  { name: "Monitor", price: 15000, quantity: 1 }
];

// Total cart value
const totalValue = cart.reduce((total, item) => {
  return total + (item.price * item.quantity);
}, 0);

// Most expensive single item (based on price)
const mostExpensive = cart.reduce((max, item) => {
  return item.price > max.price ? item : max;
});

console.log("Total Value:", totalValue);
console.log("Most Expensive Item:", mostExpensive);
```

Question 7 : Explain what array destructuring is and write code that extracts the first

Array destructuring allows us to unpack values from arrays into separate variables.

Question 9 : Write a function called calculateTotal that can accept any number of
prices as arguments and returns their sum. Use the rest operator to handle the variable
number of parameters.


```python
function calculateTotal(...prices) {
  return prices.reduce((sum, price) => sum + price, 0);
}
console.log(calculateTotal(100, 200, 300, 400));
```

The rest operator collects all arguments into an array.

Question 10 : You have two arrays of student names and need to create a new array
that contains all students from both arrays. Also, you have two objects representing a
student's basic info and marks. Write code using the spread operator to solve both
scenarios.
const classA = ["Rahul", "Priya", "Amit"];
const classB = ["Sneha", "Vikram", "Pooja"];


Merging Two Arrays (Spread Operator)

```python
const classA = ["Rahul", "Priya", "Amit"];
const classB = ["Sneha", "Vikram", "Pooja"];
const allStudents = [...classA, ...classB];
console.log(allStudents);
```

Merging Two Objects (Spread Operator)

```python
const basicInfo = { name: "Rahul", age: 20 };
const marksInfo = { maths: 90, science: 85 };
const studentDetails = { ...basicInfo, ...marksInfo };
console.log(studentDetails);
```

The spread operator  expands elements of arrays or properties of objects.
